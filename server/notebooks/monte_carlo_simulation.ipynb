```python
import json
import numpy as np
import pandas as pd
```

```python
INPUT_FILE = "../data/exports/asteroid_trajectories.json"

OUTPUT_FILE = "../data/exports/asteroid_risk_clouds.json"

N_SIMULATIONS = 1000
```

```python
with open(INPUT_FILE, "r") as f:
    data = json.load(f)

asteroids = data["asteroids"]

print("Asteroids:", len(asteroids))
```

```python
def perturb_position(x, y, z):

    x_new = np.random.normal(
        loc=x,
        scale=max(abs(x)*0.01, 0.001)
    )

    y_new = np.random.normal(
        loc=y,
        scale=max(abs(y)*0.01, 0.001)
    )

    z_new = np.random.normal(
        loc=z,
        scale=max(abs(z)*0.01, 0.001)
    )

    return x_new, y_new, z_new
```

```python
"Monte Carlo Simulation"
```

```python
risk_clouds = []
```

```python
for asteroid in asteroids:

    asteroid_id = asteroid["asteroid_id"]

    hazard_probability = asteroid["hazard_probability"]

    trajectory = asteroid["trajectory"]

    midpoint = trajectory[
        len(trajectory) // 2
    ]

    x = midpoint["position"]["x"]
    y = midpoint["position"]["y"]
    z = midpoint["position"]["z"]

    cloud = []

    for _ in range(N_SIMULATIONS):

        px, py, pz = perturb_position(
            x,
            y,
            z
        )

        distance = np.sqrt(
            px**2 +
            py**2 +
            pz**2
        )

        cloud.append({

            "x": float(px),

            "y": float(py),

            "z": float(pz),

            "distance": float(distance)

        })

    risk_clouds.append({

        "asteroid_id": asteroid_id,

        "hazard_probability":
        float(hazard_probability),

        "cloud": cloud

    })
```

```python
with open(
    OUTPUT_FILE,
    "w"
) as f:

    json.dump(
        {
            "risk_clouds": risk_clouds
        },
        f,
        indent=2
    )

print("Saved:", OUTPUT_FILE)
```
