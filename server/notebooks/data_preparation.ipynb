```python
import pandas as pd
import numpy as np
```

```python
asteroids = pd.read_csv(
    "../data/raw/near_earth_asteroids_2025.csv",
    low_memory=False
)

approaches = pd.read_csv(
    "../data/raw/asteroid_close_approaches_2015_2035.csv"
)
```

```python
asteroids["full_name"] = (
    asteroids["full_name"]
    .astype(str)
    .str.strip()
)

approaches["full_name"] = (
    approaches["full_name"]
    .astype(str)
    .str.strip()
)
```

```python
merged = approaches.merge(
    asteroids,
    on="full_name",
    how="inner"
)

print(len(merged))
```

```python
drop_cols = [
    "rot_per",
    "albedo",
    "name"
]

merged = merged.drop(
    columns=drop_cols,
    errors="ignore"
)
```

```python
merged["orbit_stretch"] = (
    merged["ad"]
    - merged["q"]
)

merged["interaction_index"] = (
    1 / (merged["moid_au"] + 1e-6)
)

merged["size_energy_index"] = (
    merged["diameter_km"] ** 2
)

merged["diameter_velocity_proxy"] = (
    merged["diameter_km"]
    * merged["velocity_km_s"]
)
```

```python
merged["close_approach_date"] = pd.to_datetime(
    merged["close_approach_date"]
)

merged["year"] = (
    merged["close_approach_date"]
    .dt.year
)

merged["month"] = (
    merged["close_approach_date"]
    .dt.month
)
```

```python
merged.to_csv(
    "../data/processed/merged_dataset.csv",
    index=False
)

merged.to_parquet(
    "../data/processed/merged_dataset.parquet",
    index=False
)
```
